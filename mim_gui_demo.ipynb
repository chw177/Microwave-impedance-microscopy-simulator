# mimsim — live MIM control panel (real-time)

A running **strip-chart**: the signal streams against a real **time axis**, and moving any slider changes the ongoing signal live. Points scrolling out of the ~12 s window are dropped automatically, so memory stays bounded no matter how long it runs.

### How to run
1. Run the two code cells below (needs `%matplotlib widget` / ipympl — you have it).
2. **Press the ▶ Play button** (top of the panel) to start streaming. Use its ⏸ / ⏹ to pause / stop.

### Four live panels
I/Q plane (trajectory) · pre-demod power vs time · MIM-Re vs time · MIM-Im vs time.

### Controls
- **Source power / Carrier freq** — scale and retune the whole chain.
- **Cancellation A / phase** — null the carrier; watch the pre-demod power dive (or click **Auto-null cancellation**).
- **Reference LO phase** — rotates loss (MIM-Re) vs capacitance (MIM-Im).
- **Sample log10 rho** — move along the response curve.
- **live noise** — adds a little instrument jitter so the stream looks alive (uncheck for clean values).
- **Sweep ρ / Sweep frequency** — draw the full response curve / frequency dip below (static, does not disturb the live stream).

_Dragging a slider updates the readouts and the current I/Q point instantly, even before the next stream tick._
